## **1.Filling the Missing Value in CorridorID**

```python

# Fill 'corridorID' based on 'corridorName'
corridor_name_mapping = df.dropna(subset=['corridorName']).drop_duplicates('corridorName')
corridor_name_mapping = corridor_name_mapping.set_index('corridorName')['corridorID']
df['corridorID'].fillna(df['corridorName'].map(corridor_name_mapping), inplace=True)
df = df.dropna(subset=['corridorID', 'corridorName'], how='all')
```

**Brief Explanation**

1. corridor_name_mapping = df.dropna(subset=['corridorName']).drop_duplicates('corridorName'): This line creates a new DataFrame corridor_name_mapping by dropping rows in df where ‘corridorName’ is NaN and then removing duplicate ‘corridorName’ rows.

2. corridor_name_mapping = corridor_name_mapping.set_index('corridorName')['corridorID']: This line sets ‘corridorName’ as the index of corridor_name_mapping and keeps only the ‘corridorID’ column. The result is a Series where the index is ‘corridorName’ and the values are ‘corridorID’.

3. df['corridorID'].fillna(df['corridorName'].map(corridor_name_mapping), inplace=True): This line fills NaN values in the ‘corridorID’ column of df with the ‘corridorID’ that corresponds to ‘corridorName’ in corridor_name_mapping.

4. df = df.dropna(subset=['corridorID', 'corridorName'], how='all'): This line drops rows in df where both ‘corridorID’ and ‘corridorName’ are NaN.
By using this code, you ensure that each ‘corridorName’ in your DataFrame has a corresponding ‘corridorID

**The reason why using Corridor Name**

Matching IDs and Names: Sometimes, if we know the route ID ('corridorID'), we can figure out the route name ('corridorName'). This helps us complete missing details between these two columns.

**Here, we fill missing 'corridorID' values based on the available 'corridorName'. If there's a 'corridorName' available for a row, we use the corresponding 'corridorID' to fill in missing values for this column.**

## **2.Filling the Missing Value in CorridorName**

```python
# Fill 'corridorName' based on 'corridorID'
df['corridorName'].fillna(df.groupby('corridorID')['corridorName'].transform('first'), inplace=True)
```

**Brief Explanation**

1. groupby('corridorID')['corridorName']: It groups the DataFrame by the 'corridorID' column and selects the 'corridorName' column for each group.

2. transform('first'): Within each group based on 'corridorID', it fills the missing values in 'corridorName' with the first non-null value found in that group. The 'first' method retrieves the first non-null value for each group.

**The reason why using Corridor ID**

1. Utilizing Existing Information: It uses the available 'corridorID' values to fill in missing 'corridorName' values. Since 'corridorID' and 'corridorName' are related, having one can often help infer the other.

2. Group-based Imputation: It employs a group-based strategy. For each unique 'corridorID', it identifies the missing 'corridorName' values and fills them based on the first non-null 'corridorName' within that 'corridorID' group. This method often captures the general information pattern within each corridor.

## **5.Filling the Missing Value in 'payAmount'**

```python
# Fill 'payAmount' based on 'corridorID' or 'tapInStops'
for column in ['payAmount']:
    df[column].fillna(df.groupby('corridorID')[column].transform('median'), inplace=True)
    df[column].fillna(df.groupby('tapInStops')[column].transform('median'), inplace=True)
    df[column].fillna(df.groupby('tapOutStops')[column].transform('median'), inplace=True)
    df = df.dropna(subset=[column, 'corridorID', 'tapInStops','tapOutStops'], how='all')
```

**The reason for grouping by ‘corridorID’, ‘tapInStops’, and ‘tapOutStops’ when filling missing values in ‘payAmount’**

1. ‘corridorID’: This might represent a specific route or path that the transportation system follows. The ‘payAmount’ could be related to the distance or complexity of the corridor. Grouping by ‘corridorID’ assumes that similar ‘payAmount’ values are more likely to occur within the same corridor.

2. ‘tapInStops’: This could be the station where passengers get on the transportation system. The ‘payAmount’ might be influenced by the starting point of the journey. Grouping by ‘tapInStops’ assumes that similar ‘payAmount’ values are more likely to occur for journeys starting from the same stop.

3. ‘tapOutStops’: This could be the station where passengers get off the transportation system. The ‘payAmount’ might be influenced by the end point of the journey. Grouping by ‘tapOutStops’ assumes that similar ‘payAmount’ values are more likely to occur for journeys ending at the same stop.

**‘PayAmount’ has outliers or is not normally distributed, using the median would be a more appropriate choice for imputation.
```python
from sklearn.impute import KNNImputer

# Assuming df is your DataFrame

# Selecting columns for KNN imputation
columns_to_impute = ['tapOutStopsLat', 'tapOutStopsLon']

# Creating a KNNImputer instance with the desired number of neighbors
knn_imputer = KNNImputer(n_neighbors=5)

# Filling missing values using KNN imputer
df[columns_to_impute] = knn_imputer.fit_transform(df[columns_to_impute])
```

## **9. Filling the Missing Value in stopEndSeq**

```python
# Sort the DataFrame based on tapOutStopsName or the sequence order
df = df.sort_values(by='tapOutStopsName')

# Interpolate missing values in stopEndSeq based on tapOutStopsName sequence
df['stopEndSeq'] = df['stopEndSeq'].interpolate(method='linear', limit_direction='both')

# 'linear' method fills missing values linearly, and 'limit_direction' ensures interpolation in both directions
```

**Brief Explanation**

1. Sorting based on tapOutStopsName: Sorting the DataFrame based on the tapOutStopsName column ensures that the sequence of stops is in the correct order. This step is crucial as interpolation will be based on this sorted sequence.

2. Interpolation with 'linear' method: The interpolate() method with method='linear' fills missing values in a linear manner. When dealing with sequences such as stop numbers along a route, linear interpolation assumes a uniform progression between known values. This method is suitable when the stops have a consistent order and spacing between them.

3. limit_direction='both': Setting limit_direction='both' ensures that interpolation occurs in both directions from the available data. If there are consecutive missing values, this parameter allows the interpolation to happen from the closest known values both forward and backward in the sequence.

4. Handling missing stopEndSeq values: By performing interpolation based on the sorted sequence of tapOutStopsName, the code helps to estimate and fill in missing stopEndSeq values, thereby preserving the sequence of stops along the route.

**This approach assumes that the tapOutStopsName represents a sequential order of stops and that the missing stopEndSeq values can be reasonably estimated based on this sequence. It helps maintain the integrity of the route sequence by ensuring the stopEndSeq values follow a logical progression along the stops, which is often essential for transportation or route-related data.**

**For the 'stopEndSeq' column, considering it represents a sequence of stop endings, imputing missing values might require a bit of contextual understanding. Here are a few potential strategies:**

1. Mean/Median/Mode Imputation: Impute missing 'stopEndSeq' values with the mean, median, or mode of the existing values in that column. This method assumes the missing values can be estimated based on the central tendency of the available data.

2. Sequence-Based Imputation: If there's a known sequence pattern in the data, you might fill missing values based on this sequence. For instance, if stop sequences follow a particular order or increment by a consistent amount, you could infer missing values based on this pattern.

3. Grouping by Relevant Features: If there are other features (such as 'corridorID', 'tapOutStops', or 'tapOutTime') that relate to 'stopEndSeq', you can group the data by these features and impute missing values based on the patterns within each group.

4. The recommendation to use interpolation for filling missing values in stopEndSeq might stem from the nature of the data and its sequential order. In transportation or routing data, stopEndSeq often represents the sequence number of a stop along a route or journey. If there are missing values in this sequence, it might disrupt the continuity of the route information or cause issues when analyzing the sequential order of stops.

Interpolating missing stopEndSeq values can help maintain the sequential order of stops along the route. Since this data represents a sequence of stops, using interpolation methods like linear or cubic can estimate the missing sequence numbers based on the known sequence numbers before and after the missing values.

```python

# Selecting numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Creating boxplots for numerical columns arranged in a 2x5 grid
fig, axes = plt.subplots(2, 5, figsize=(20, 10))

for i, column in enumerate(numerical_columns):
    sns.boxplot(x=df[column], ax=axes[i // 5, i % 5])
    axes[i // 5, i % 5].set_title(f'Boxplot of {column}')
    axes[i // 5, i % 5].set_xlabel(column)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Dictionary to store outliers' information
outliers_info = {}

for column in numerical_columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers_condition = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers = df[outliers_condition]

    # Store information in the dictionary
    outliers_info[column] = {
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': outliers_condition.sum(),  # Count of outliers
        'outliers_indices': outliers.index.tolist()  # Indices of outliers
    }


# summary for outliers_summary

outliers_summary = {key: {'bounds': (value['lower_bound'], value['upper_bound']), 
                          'number_of_outliers': value['number_of_outliers']} 
                    for key, value in outliers_info.items()}

# Creating DataFrame from 'outliers_summary'
outliers_summary_df = pd.DataFrame.from_dict(outliers_summary, orient='index')

# Splitting 'bounds' into 'lower_bound' and 'upper_bound' columns
outliers_summary_df[['lower_bound', 'upper_bound']] = outliers_summary_df['bounds'].apply(pd.Series)

# Dropping the original 'bounds' column
outliers_summary_df.drop('bounds', axis=1, inplace=True)

# Reordering the columns for a better view
outliers_summary_df = outliers_summary_df[['lower_bound', 'upper_bound', 'number_of_outliers']]

# Display DataFrame as a styled table
styled_table = outliers_summary_df.style.format('{:.0f}')
styled_table
```

```python
def create_age_category(cols):
    current_year = pd.Timestamp.now().year
    age = current_year - cols.year  # Calculate age based on the year of birth

    if age <= 5:
        return 'Toddler/Preschooler' 
    elif 5 < age <= 12:
        return 'Child'  
    elif 12 < age <= 18:
        return 'Adolescent/Teenager' 
    elif 18 < age <= 60:
        return 'Adult'  
    else:
        return 'Elderly/Senior'  

# Applying the age category function to create a new 'ageCategory' column
df['ageCategory'] = df['payCardBirthDate'].apply(create_age_category)

df.head()

```

## **Morning and Evening Rush**

```python
# Function to categorize rush hours
def categorize_rush_hour(hour):
    if 6 <= hour < 10:
        return 'Morning Rush Hour'  # 6 AM to 9:59 AM
    elif 17 <= hour < 20:
        return 'Evening Rush Hour'  # 5 PM to 7:59 PM
    else:
        return 'No Rush Hour'

# Apply the rush hour categorization function to create a new 'rushHour' column
df['rushHour'] = df['tapInHour'].apply(categorize_rush_hour)

df.head()
```

1. payCardID and Other Variables: There is a very weak negative correlation with all other variables, indicating almost no linear relationship. This suggests that the ID of the pay card does not significantly influence other factors such as direction, stop sequence, or payment amount.

2. direction: It has no correlation with itself (as expected) and very weak correlations with other variables, suggesting that direction does not have a significant linear relationship with them. This means the direction of travel does not significantly affect other factors.

3. tapinStopsLat and tapinStopsLon: These two have a weak positive correlation of 0.03, indicating a slight linear relationship between the latitude and longitude where passengers tap in. This suggests that certain geographical areas may see more tap-ins.

4. tapoutStopsLat and tapoutStopsLon: A strong positive correlation of 0.91 indicates that as the latitude of where passengers tap out increases, so does the longitude, suggesting these events occur in proximity. This could indicate that certain areas have higher tap-out rates.

5. stopStartSeq and stopEndSeq: A moderate positive correlation of 0.79 suggests a notable linear relationship; as the starting sequence number of stops increases, so does the ending sequence number. This could indicate that longer routes have more stops.

```python
numerical_col = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_col:
    corr = df[col].corr(df["payAmount"], method="spearman")

    if corr > 0:
        correlation = "positive"
        if  corr > 0.0 and corr <= 0.2:
            strength = "very weak"
        elif corr > 0.2 and corr <= 0.4:
            strength = "weak"    
        elif corr > 0.4 and corr <= 0.6:
            strength = "moderate"
        elif corr > 0.6 and corr <= 0.9:
            strength = "strong"
        else:
            strength = "very strong"
    else:
        correlation = "negative"
        if corr  < 0.0 and corr >= -0.2:
            strength = "very weak"
        elif corr < -0.2 and corr >= -0.4:
            strength = "weak"    
        elif corr < -0.4 and corr >= -0.6:
            strength = "moderate"
        elif corr < -0.6 and corr >= -0.9:
            strength = "strong"
        else:
            strength = "very strong"

    print(f"The correlation value between the {col} and PayAmount columns is {corr:.2f}")
    print(f"The relationship between the {col} and PayAmount variables is '{correlation}' with a correlation strength of '{strength}'.")
    print()
```
